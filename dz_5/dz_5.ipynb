**1 Задание**

**Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16.
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95, если выборочная средняя M = 80, а объем выборки n = 256.**

$\sigma=16$

$M(X)=a$

$P = 0.95$

$\overline{X_в}=80$

$n=256$

уже был выполнен анализ генеральной совокупности $\Rightarrow$ критерием оценки $K$ будет критерий Фишера $Z$

из таблицы распределения Лапласа для $Z<0$

генеральная совокупность распределена нормально и у нас задача на доверительный интервал $\Rightarrow$ мы можем от критерия 

$\alpha$ перейти к $\frac{\alpha}{2}$

$\alpha=1-P=1-0.95=0.05$

$\frac{\alpha}{2}=\frac{0.05}{2}=0.025$

так как был выполнен анализ генеральной совокупности $\Rightarrow$ стандартная ошибка среднего $=\frac{\sigma}{n}$, где

$\sigma$ взято из генеральной совокупности

$Z(P_z=0.025)=-1.96$

находим доверительный интервал по формуле:

$\left[
  \begin{array}{ccc}
     \overline{X_в} \pm Z_{\frac{\alpha}{2}}\cdot \frac{\sigma}{\sqrt{n}}\\
  \end{array}
\right]
\Rightarrow
\left[
  \begin{array}{ccc}
     80 \pm 1.96\cdot \frac{16}{\sqrt{256}}\\
  \end{array}
\right]
\Rightarrow
\left[
  \begin{array}{ccc}
     80 - 1.96; 80 + 1.96\\
  \end{array}
\right]
$

$\left[
  \begin{array}{ccc}
     78.04; 81.90\\
  \end{array}
\right]
-$ этот интервал с вероятностью 95% покрывает истинное математическое ожидание $M(X)=a$ генеральной совокупности

**2 Задание**

**В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.**

неизвестно $\sigma \Rightarrow$ не было аналищв генеральной совокупности $\Rightarrow$ критерием оценки будет критерий Стьюдента $t$ из таблицы Стьюдента

Тогда стандартная ошибка это $\frac{s}{\sqrt{n}}$, где $s$ - это корень из дисперсии по выборке $s=\sqrt{D_в(X)}$, 

где $\sqrt{D_в(X)}$ - несмещённая оценка дисперсии

испытания проводились с одинаковой точностью $\Rightarrow$ генеральная совокупность распределена нормально $\Rightarrow$ мы можем перейти от $\alpha$ к $\frac{\alpha}{2}$

$P=0.95 \Rightarrow \alpha=1-P=1-0.95=0.05$

$\frac{\alpha}{2} = \frac{0.05}{2}=0.025$

находим среднюю величину по выборке $\overline{X_в}$

$\overline{X_в}=\frac{6.9+6.1+6.2+6.8+7.5+6.3+6.4+6.9+6.7+6.1}{10}=6.59$

находим несмещённую оценку дисперсии

$s^2=\frac{1}{n-1}\sum\limits_{i=1}^{n} (x_i-\overline{x_в})^2 =$

$=\frac{(6.9-6.59)^2+(6.1-6.59)^2+(6.2-6.59)^2+(6.8-6.59)^2+(7.5-6.59)^2+(6.3-6.59)^2+(6.4-6.59)^2+(6.9-6.59)^2+(6.7-6.59)^2+(6.1-6.59)^2}{10-1}=$

$=\frac{0.0961+0.2401+0.1521+0.0441+0.8281+0.0841+0.0361+0.0961+0.0121+0.2401}{9}=0.2032$

$s=\sqrt{0.2032}=0.4508$

критерий Стьюдента $t=t_{\frac{\alpha}{2}} = 2.262$

находим доверительный интервал по формуле:

$\left[
  \begin{array}{ccc}
     \overline{X_в} \pm t_{\frac{\alpha}{2}}\cdot \frac{s}{\sqrt{n}}\\
  \end{array}
\right]
\Rightarrow
\left[
  \begin{array}{ccc}
     6.59 \pm 2.262\cdot \frac{0.4508}{\sqrt{10}}\\
  \end{array}
\right]
\Rightarrow
\left[
  \begin{array}{ccc}
     6.59 - 0.3225; 6.59 + 0.3225\\
  \end{array}
\right]
$

$\left[
  \begin{array}{ccc}
     6.2675; 6.9125\\
  \end{array}
\right]
-$ этот интервал покрывает истинное значение $X$ с доверительной вероятностью 95%

**3 Задание**

**Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.**

$\mu_0=17$

$\mu=17.5$

$n=100$

$\alpha=0.05$

$D=4$

Гиптоеза:

$H_0: \mu = \mu_0$

$H_1: \mu > \mu_0$

известна дисперсия $\Rightarrow$ был анализ генеральной совокупности

$\sigma=\sqrt{D(X)}=\sqrt{4}=2$

стандартная ошибка среднего = $\frac{\sigma}{\sqrt{n}}$

так как был анализ генеральной совокупности, то критерием оценки будет критерий Фишера $Z$ по таблице для $Z>0$

найдём и сравним $Z_{табл}$ и $Z_{расч}$

$Z_{табл}=Z(1-\alpha)=Z(1-0.05)=Z(P=0.95)=1.65$

$Z_{расч}=\frac{\mu-\mu_0}{\frac{\sigma}{\sqrt{n}}}=\frac{17.5-17}{\frac{2}{10}}=\frac{5}{2}=2.5$

$Z_{расч} > Z_{табл} \Rightarrow$ отвергаем гипотезу $H_0$, гипотеза $H_1$ на уровне значимости $5\%$ верна и $\mu=17.5$

**4 Задание**

**Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?**

$n=10$

$\alpha=1\%$

$P=0.99$

$\mu_0=200$

$\mu=\overline{X_в}$

Гиптоеза:

$H_0: \mu = \mu_0$

$H_1: \mu \neq \mu_0$

неизвестно $\sigma \Rightarrow$ не было анализа генеральной совокупности $\Rightarrow$ за критерий оценки возьмём критерий Стьюдента $t$

стандартная ошибка среднего $=\frac{s}{\sqrt{n}}$, где $s$ - среднеквадратичное отклонение от несмещённой оценки дисперсии выборки

находим среднюю величину по выборке $\overline{X_в}$

$\overline{X_в}=\frac{202+203+199+197+195+201+200+204+194+190}{10}=198.5$

находим несмещённую оценку дисперсии

$s^2=\frac{1}{n-1}\sum\limits_{i=1}^{n} (x_i-\overline{x_в})^2 =$

$=\frac{(202-198.5)^2+(203-198.5)^2+(199-198.5)^2+(197-198.5)^2+(195-198.5)^2+(201-198.5)^2+(200-198.5)^2+(204-198.5)^2+(194-198.5)^2+(190-198.5)^2+}{10-1}=$

$=\frac{12.25+20.25+0.25+2.25+12.25+6.25+2.25+30.25+20.25+72.25}{9}=\frac{178.5}{9}=19.83$

$s=\sqrt{19.83}=4.4535$

найдём и сравним $t_{табл}$ и $t_{расч}$

*для табличного:*

$k=n-1=10-1=9$

$p=1-\alpha=1-0.01=0.99$

$\Rightarrow t_{табл} = 2.821$

*для расчётного:*

$t_{расч} = \frac{|\mu-\mu_0|}{\frac{s}{\sqrt{n}}}=\frac{|198.5-200|}{\frac{4.4535}{\sqrt{10}}} = \frac{1.5\cdot\sqrt{10}}{4.4535}= \frac{4.4734}{4.4535}=1.065$


$t_{расч} < t_{табл} \Rightarrow$ принимаем гипотезу $H_0$, согласно которой $\mu=200$ на уровне значимости $99\%$
